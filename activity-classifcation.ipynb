```python
from keras.models import Sequential
from keras.layers import Input,Convolution2D,MaxPooling2D,Flatten,Dense,Dropout
from keras.utils import np_utils

import tensorflow

```

```python
model1 = Sequential()
model1.add(Convolution2D(32, (11,11), activation='relu',input_shape=(128,128,3),strides=(2,2)))
model1.add(MaxPooling2D(pool_size=(3,3),strides=(2,2)))
# model1.add(Dropout(0.1))
model1.add(Convolution2D(64, (5,5), activation='relu'))
model1.add(MaxPooling2D(2,2))
# model1.add(Dropout(0.25))
model1.add(Convolution2D(128,(3,3),activation='relu'))
#model1.add(MaxPooling2D(2,2))
model1.add(Convolution2D(256,(3,3),activation='relu'))
model1.add(Dropout(0.5))
#model1.add(Convolution2D(512,(3,3),activation='relu'))
#model1.add(Dropout(0.3))
model1.add(Flatten())
model1.add(Dense(500,activation='relu'))
model1.add(Dropout(0.3))
model1.add(Dense(100,activation='relu'))
model1.add(Dropout(0.3))
model1.add(Dense(21,activation='softmax'))

model1.summary()
```

```python
checkpoint=ModelCheckpoint('./base.model',
                          monitor='val_acc',
                          verbose=1,
                          save_best_only=True,
                          mode='max',
                          save_weights_only=False,
                          period=1)
earlystop=EarlyStopping(monitor='val_acc',
                       min_delta=0.000001,
                       patience=30,
                       verbose=1,
                       mode='auto')

tensorboard = TensorBoard(
    log_dir = './logs',
    histogram_freq=0,
    batch_size=16,
    write_graph=True,
    write_grads=True,
    write_images=False,
)

csvlogger = CSVLogger(
    filename= "training_csv.log",
    separator = ",",
    append = False
)

lrsched = LearningRateScheduler(step_decay,verbose=1)

reduce = ReduceLROnPlateau(
    monitor='val_acc',
    factor=0.8,
    patience=5,
    verbose=1, 
    mode='auto',
    min_delta=0.0001, 
    cooldown=1, 
    min_lr=0.0001
)
```

```python
callbacks = [checkpoint,tensorboard,earlystop,csvlogger,reduce]
```

```python


model1.compile(loss='categorical_crossentropy',optimizer='rmsprop',metrics=['accuracy'])
```
